## Abstract

This project explored the research question: To what extent do headshot percentage and agent pick rates predict team win rates across regions in the 2024-2025 VCT playoffs, and does the strength of these predictors vary significantly between regions? Understanding what drives success in the professional scene of Valorant is valuable to fans (Pick-Ems), analysts, and teams seeking competitive advantages. 

Using a Kaggle dataset that web-scraped data from vlr.gg, we aggregated player statistics with team statistics and combined them with agent pick rate data. We created multiple visualizations, such as scatterplots of headshot percentages versus win rate, regional boxplots comparing headshot and win rate distributions, and agent pick rate bar charts. We also tested our hypothesis on whether headshot rate correlated with win rates and how they varied by region and by agent pick rate by creating a multilinear regression model and doing PCA analysis. 

Our EDA analysis revealed several interesting findings. First, it revealed that headshot percentage alone is not a good predictor of team success. The scatterplot overall is a weak and inconsistent predictor of team success. The scatterplot showed no clear positive correlation, and some of the highest headshots actually had some of the lowest win rates. Second, agent selection is driven by strategic utility rather than headshot percentage. All the top 10 agents showed similar headshot distributions with their median often hovering around 20-30%. Third, the relationship between these predictors and win rates varies drastically by region. Americas and Pacific both showed the most consistent headshot percentages, but Americas had the most inconsistent win rates, while Pacific had the most consistent results. Meanwhile, EMEA had the widest headshot variability yet maintained the highest median win rate. Additionally, the boxplots relating to China revealed a higher median headshot rate than Americas and Pacific, but varied slightly less than Americas regarding win rate despite having a few select outliers. Our findings suggest that headshot percentage and agent pick rates are unreliable as universal predictors of win rates, and that regional context is crucial for evaluating the outcome of professional matches. 

Our multilinear regression model yielded an R<sup>2</sup> value of 0.786 and p-value of 0.039, meaning that about 78.6% of the variation in win rate is explained by the model and that headshot rate is statistically significant in predicting win rates with a negative coefficient of -2.44 respectively. These findings aligned with the EDA scatterplot, which revealed a negative correlation between headshot rate and win rate. As for PCA, we found that 49.92% of the total variance in the dataset indicated how the first two principal components represented a large portion of the variation in agent pick compositions and mechanical performance across matches in a two-dimensional space. This helped us conclude that higher headshot rate, while significantly significant, does not necessarily guarantee a team having a higher win rate nor does their agent pick rates or region.

## Background and Prior Work

### Introduction

Developed and released in 2020 by Riot Games, VALORANT is a free-to-play 5v5 tactical shooter game with 28 playable agents divided into 4 different roles–duelist, initiator, controller, sentinel–as of February 2026. Players select “agents” prior to the start of the match, with teams switching between attacking and defending “sites” halfway through the game. “Site” refers to an area on the map where the attacking team needs to plant the objective, otherwise known as the “spike,” and the defending team needs to prevent it from being planted or defuse it before it explodes after a certain amount of time. “Agents” are playable characters, each with unique abilities that complement the different roles: 
- Duelists: Excel at “entry fragging.” Basically, they are agents that are best known for initial fights and entering sites. Their abilities are often movement abilities or defensive utilities meant for getting kills to help their team enter the site and plant the bomb.
- Initiators: Information gatherers. These agents' utilities help their team find where the enemies are positioning or even hiding. Additionally, their kit comes with abilities that help the team further fight enemies at an advantage, such as flashes or stuns. Flashing refers to blinding enemies, and stunning refers to a short debuff that impairs movement, gunfire rate, and shaky camera movement for the player to simulate dizziness. Both of these factors may increase a player’s probability of winning a gunfight with their opponent.
- Controllers: Strategic agents that utilize their abilities to assist teammates and manipulate the battlefield, which involves the map, enemy positions, and actions. These abilities include walls and smokes, which help block out enemies and help their team take space on site. Walls and smokes are temporary barriers that block the enemy’s line of sight, but not a solid barrier, so they can still walk through them or enemies can also wait for them to disappear after a certain amount of time.
- Sentinels: Site defenders and bombsite anchors. Sentinels are the agents who help secure the sites for their team. They excel at stalling enemy pushes–whether it be delaying an enemy attack from behind or delaying the enemy team from taking full control of a site.

The goal of the game is to win 13 rounds by attacking and defending your site using gunplay and abilities. A core mechanic in gunplay is the headshot. While each gun deals a different amount of damage to the head, something unique to Valorant is that when both teams are on equal buy, meaning they have the same amount of money, most guns only require a single bullet to the head to instantly kill an opponent. This makes headshot percentage a key measurement of individual mechanical skill. 

VALORANT Champion Tour (VCT) is an annual tournament hosted by Riot Games. It features four regional leagues, Americas, EMEA, Pacific, and China. VCT events are divided into regional leagues and international leagues. Regional leagues refer to the four regions as mentioned previously while international leagues refer to Masters or Champions events. In other words, the international leagues consist of top teams that won their regional games and the one team that wins is considered the best team in the world as of that year. Since there are many different levels of tournaments and stages, for this project, we will be focusing on VCT Challengers playoff games from 2024. We decided to leave out games from previous stages since the playoffs stage involves top teams, where winners advance to the grand finals.

From a viewer's perspective, each region has different play styles. Regions like EMEA may play very strategically while other regions like to ‘run it down,’ or in other words, play more aggressively. With 18 unique guns to buy from and 28 different agents to choose from, we thought it would be interesting to explore how headshot rates and agent pick rates affect win percentages across different regions. 

This research matters because VCT playoffs represent the highest level of competitive Valorant. The prize pools for these international events exceed a million dollars annually. With any kind of sport, successful teams will attract huge sponsorship deals and brand partners. Understanding certain features that establish a successful team can help improve future games and coaching decisions. It might even shine light on any adjustments necessary for improving game balance by showing us whether certain playstyles or agents have a possible disproportionate competitive advantage. In addition, regional playstyle differences may reveal how gameplay is tied to cultural differences and practice environments.

```python
# create function that converts percent strings into float values

def convert_percent_column(series, as_proportion=False):
    """
    purpose: converts pandas Series like '64%' into numeric
    
    params:
        series (pd.Series): column containing percentage strings
        as_proportion (bool): if True, returns proportion as a decimal (e.g 0.64 instead of 64)
    
    returns:
        pd.Series (float)
    """
    cleaned_series = (series.astype(str).str.strip().str.replace('%', '', regex = False))
    
    # convert
    numeric = pd.to_numeric(cleaned_series, errors='coerce')  
    
    if as_proportion:
        numeric = numeric / 100
        
    return numeric
```

```python
# convert percentages column from object into float data types
overview_clean['Headshot %'] = convert_percent_column(
    overview_clean['Headshot %'],
    as_proportion = True
)

# since we will eventually look into agent pick rate, we will also convert the 'Pick Rate' column here
agents_pick['Pick Rate'] = convert_percent_column(
    agents_pick['Pick Rate'],
    as_proportion = True
)
```

```python
# calculate headshot percentage as a team
playoffs_overview = overview_clean[overview_clean['Stage'] == 'Playoffs']

playoffs_overview = (
    playoffs_overview
    .groupby('Team')['Headshot %']
    .mean()
    .reset_index(name = 'avg_team_hs')
)

print(playoffs_overview.shape)
playoffs_overview.head()
```

Now that we have calculated the average headshot rate per team, we need to calculate each team's win rate of a match. We will need the ``scores`` dataframe to calculate this.

```python
# calculate team win rate

# make long dataframe (one row per team per match)
playoffs_scores = scores[scores['Stage'] == 'Playoffs']
df = playoffs_scores.copy()

team_a = df[[
    "Tournament", "Stage", "Match Type",
    "Team A", "Match Result"
]].copy()

team_a.columns = [
    "Tournament", "Stage", "Match Type",
    "Team", "Match Result"
]

team_b = df[[
    "Tournament", "Stage", "Match Type",
    "Team B", "Match Result"
]].copy()

team_b.columns = [
    "Tournament", "Stage", "Match Type",
    "Team", "Match Result"
]

matches_long = pd.concat([team_a, team_b], ignore_index=True)

# determine who won the match
matches_long["Winner"] = matches_long["Match Result"].str.replace(
    " won", "", regex = False
)

matches_long["Win"] = (
    matches_long["Team"] == matches_long["Winner"]
).astype(int)

# aggregate to team level
team_record = (
    matches_long
    .groupby("Team")
    .agg(
        Matches_Played = ("Win", "count"),
        Wins = ("Win", "sum")
    )
    .reset_index()
)

team_record["Losses"] = (
    team_record["Matches_Played"] - team_record["Wins"]
)

team_record["Win_Rate"] = (
    team_record["Wins"] / team_record["Matches_Played"]
)

team_record = team_record.sort_values("Win_Rate", ascending = False)

print(team_record.shape)
team_record.head()
```

Now, we can merge these dataframes together since it's easier to have everything in one place. But we are not done yet! We also need to calculate the agent pick rate for each team.

```python
# merge what we have so far on team level (playoffs_overview and team_record)
playoffs_teams_merged = pd.merge(playoffs_overview, team_record, on = 'Team', how = 'inner')

print(playoffs_teams_merged.shape)
playoffs_teams_merged.head()
```

```python
teams_agents_pick.head()
```

```python
# calculate average pick rate of agents for each team in playoffs stage

'''
procedure:
1) filter teams_agents_pick by playoffs stage
2) groupby team + agent & calc avg pick rate for each team-agent combo
3) pivot table to have teams as rows and agents as columns, w avg pick rate
'''

# filter by playoffs stage
teams_agents_pick = teams_agents_pick[teams_agents_pick['Stage'] == 'Playoffs']

# count agent picks
count_agent_picks = teams_agents_pick.groupby(["Team","Agent"]).size().reset_index(name="agent_picks")

# total maps per team
count_team_maps = teams_agents_pick.groupby("Team")["Map"].nunique().reset_index(name="total_maps")

# merge
playoffs_team_agent_picks = count_agent_picks.merge(count_team_maps, on="Team")

# calculate pick rate
playoffs_team_agent_picks["pick_rate"] = playoffs_team_agent_picks["agent_picks"] / playoffs_team_agent_picks["total_maps"]
playoffs_team_agent_picks.head()
```

```python
# pivot table to have teams as rows and agents as columns, w avg pick rate

pivoted = playoffs_team_agent_picks.pivot(index="Team", columns="Agent", values="pick_rate").reset_index()
pivoted = pivoted.fillna(0.0)
pivoted.columns = [
    col if col == "Team" or col == "Agent" else f"{col}_pick_rate"
    for col in pivoted.columns
]

print(pivoted.shape)
pivoted.head()
```

```python
# merge w playoffs_teams_merged to have final dataframe with all team-level stats + agent pick rates
final_df = pd.merge(playoffs_teams_merged, pivoted, on = 'Team', how = 'left')
final_df = final_df.rename(columns = str.lower)

# only get columns we need
final_df = final_df.drop(columns = ['matches_played', 'wins', 'losses'])
```

### Multilinear Regression Analysis

As mentioned in our Wrangling section, we will be using ``final_df`` for statistical modeling and testing.

We prepared our dataframe for modeling by dropping the team name column since it's a unique identifier and would not be useful in the prediction modeling we're about to do. We also converted region names into dummy variables so that categories are binary encoded (i.e values of 0 or 1). We also used the parameter ``drop_first`` when creating these dummies. The purpose of this is to avoid perfect multicollinearity. When we create the model, our multilinear regression formula will use the Americas region as the baseline, meaning that there will still be a bias intercept to predict the win rate for America teams even if the data for it is not present in the dataframe itself.

```python
# prepare data for modeling by dropping team col since it's an identifier & not useful for modeling
df_model = final_df.drop(columns = ['team'])

# convert region into dummy vars for modeling
df_model = pd.get_dummies(df_model, columns = ['team_region'], drop_first = True)
df_model.head()
```

From a previous analysis attempt using the dataframe that was created above (and which we also omitted from the notebook), we found that putting all columns for each agent yielded insignificant results (R-squared value of 1.000, infinite standard errors) and the model overfit. This result made sense since we were using more than 20 variables to compute win rate. To combat this, we decided to group agents by their role and run multilinear regression that way.

```python
# group agents by their role & sum pick rates to get overall pick rate for each role, then add as new columns to new modeling dataframe
controllers = ['astra_pick_rate', 'brimstone_pick_rate', 'clove_pick_rate', 'omen_pick_rate', 'viper_pick_rate']
duelists = ['jett_pick_rate', 'raze_pick_rate', 'reyna_pick_rate', 'phoenix_pick_rate', 'neon_pick_rate', 'yoru_pick_rate', 'iso_pick_rate']
initiators = ['sova_pick_rate', 'skye_pick_rate', 'fade_pick_rate', 'gekko_pick_rate', 'breach_pick_rate', 'kayo_pick_rate']
sentinels = ['cypher_pick_rate', 'killjoy_pick_rate', 'sage_pick_rate', 'chamber_pick_rate', 'deadlock_pick_rate']

# create new dataframe for modeling
df_model2 = df_model[['avg_team_hs', 'win_rate', 'team_region_China', 'team_region_EMEA', 'team_region_Pacific']].copy()
df_model2['controller_pick_rate'] = df_model[controllers].sum(axis=1)
df_model2['duelist_pick_rate'] = df_model[duelists].sum(axis=1)
df_model2['initiator_pick_rate'] = df_model[initiators].sum(axis=1)
df_model2['sentinel_pick_rate'] = df_model[sentinels].sum(axis=1)

df_model2.head()
```

```python
# define X & y variables
X = df_model2.drop(columns = ['win_rate'])
y = df_model2['win_rate']

# ensure X is all float data types
X = X.astype(float)
X = sm.add_constant(X)

# fit OLS regression model & print summary
model = sm.OLS(y, X).fit()
print(model.summary())
```

#### Multilinear Regression Results Breakdown

As opposed to the other analysis attempt we described before creating this new modeling dataframe, the model summary here generated more meaningful results.

Firstly, the model generated an R<sup>2</sup> value of 0.786 and an adjusted R<sup>2</sup> of 0.701. This means about 78% of the variance in team win rates can be explained by the model, and even after adjusting for the number of predictors, about 70% of the variance in team win rates can be explained by the model.

In regards to our hypothesis, the model yielded a p-value of 0.039 which is less than our significance level of α = 0.05. This means that headshot rate is statistically significant in predicting win rates. However, it is also important to address the coefficient of average team headshot rates which is -2.43. A negative coefficient in this case means that, interestingly enough, an increase in headshot rate is associated with a decrease in win rate. This goes directly against our hypothesis that an increase in headshot would positively correlate with win rate.

Additionally, when considering the p-values we got for each region, we can see that all region variables have a p-value greater than 0.05.
- China p-value: 0.543 > 0.05
- EMEA p-value: 0.698 > 0.05
- Pacific p-value: 0.718 > 0.05

In essence, there is no statistically significant difference in win rate between regions for the VCT 2024 playoffs dataset we generated. This also goes against our hypothesis in that regional playstyles may affect win outcomes.

When looking at agent role result breakdowns, we found there to be no statistically significant evidence that team composition predicts win rate for this model given that the p-values for each role category were greater than our significance level of α = 0.05.
- Controller p-value: 0.471
- Duelist p-value: 0.469
- Initiator p-value: 0.335
- Sentinel p-value: 0.843

### Principal Component Analysis (PCA)

Based on our previous analysis, we will apply PCA to see how mechanical performance relates to agent compositions in VCT matches. PCA helps reduce the dimensionality of the data and allows us to visualize potential composition patterns across matches.

Before fitting our data to the model, we will standardize our data and keep the first 2 principal components in a separate dataframe, along with ``win_rate`` and ``team_region``. From there, we can see how much each component explains the variance between agent selection and headshot rate.

```python
# use all agents and avg team headshot rate as features for modeling
features = controllers + duelists + initiators + sentinels + ['avg_team_hs']
cols = df_model[features]

# scale the data and keep the first two principal components
scaler = StandardScaler()
features_scaled = scaler.fit_transform(cols)

pca = PCA(n_components = 2)
pcs = pca.fit_transform(features_scaled)

# create dataframe with PC1 and PC2
pca_df = pd.DataFrame(pcs, columns=['PC1', 'PC2'])

# keep win rate & region
pca_df['win_rate'] = df_model['win_rate']
pca_df['team_region'] = final_df['team_region']
pca_df = pca_df.dropna(subset = ['team_region']) # drop rows with missing region values

# show how much variance 
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

The explained variance ratio shows how much information is captured by each principal component. In our results, the first two principal components explain 37.89% and 12.03% of the variance respectively. Together, they around 49.92% of the total variance in the dataset. This indicates that these two components represent a large portion of the variation in agent pick compositions and mechanical performance across matches in a two-dimensional space.

The first component represents variation across teams while the second component captures differences in playstyle, which supports the idea that team win rate may be affected by compositional patterns instead of individual performance alone.

```python
# get PCA loadings
loadings = pca.components_.T

loading_df = pd.DataFrame(
    loadings,
    index=features,
    columns=['PC1', 'PC2']
)

# loading_df.head()
```

```python
print('PC1 results\n-----\nHighest loadings:')
print(loading_df['PC1'].sort_values(ascending = False).head(10))
print('\nLowest loadings:')
print(loading_df['PC1'].sort_values(ascending = False).tail(10))
```

Our loadings dataframe reveals that teams with a high PC1 value typically use agents that are in the meta and have a more balanced composition. Notice how headshot rate does not seem to appear in the top 10 or bottom 10 rows of our loadings, meaning that our headshot rate value is closer to 0. This means that mechanics are not the main variation between teams in this case.

```python
print('PC2 results\n-----\nHighest loadings:')
print(loading_df['PC2'].sort_values(ascending = False).head(10))
print('\nLowest loadings:')
print(loading_df['PC2'].sort_values(ascending = False).tail(10))
```

Our second PCA shows how headshot rate may be associated with agents who excel more at info-gathering and have a structured utility kit (e.g Cypher, Sova, Deadlock, Astra), which goes against our expectation that higher headshot rates are associated with duelists or more aggressive agents (e.g Jett, Chamber, Phoenix).

Additionally, we created different visualizations to help us explore the distribution of our variables and accompanied them with machine learning models such as multilinear regression and principle component analysis (PCA).

### What We Learned

In the early stages of our modelling process, we created a model that involved all variables we needed to answer our question. This meant that each team had the average pick rate for every single agent, leaving us with a dataframe shape of 29 rows and 28 columns. To create meaningful insights for our data, we grouped agents by their respective role to reduce model overfitting, then fit them into our multilinear regression and PCA models.

To reiterate, through multilinear regression, we found that headshot rate is statistically significant in predicting team win rates with a negative coefficient of -2.4348 while agent pick and region were not as statistically significant in comparison. This is supported by our first EDA scatterplot that suggested a weak, negative correlation between a team's average headshot rate and their match win rate. Additionally, our PCA results suggested that there is a trade-off between mechanics and agent compositions. PC1 values were more associated with balanced teams while PC2 values suggested a negative association between headshot rate and agents with structural or info-gathering utility. In conjunction with the scatterplot comparing between PC1 and PC2 values, we were able to see the distributions in team composition between regions. More importantly, the PCA scatterplot showed that teams from the same region did not cluster together. These results aligned with our regression results which found that region is not statistically significant in predicting team win rates. Overall, our findings support the idea that headshot rate, while statistically significant, does not necessarily guarantee higher team win rates.

### Limitations

There are several limitations that may have affected our ability to fully answer our research questions. Firstly, sample size is limited to the 24 teams (8 teams per the 4 regions) that make it to regional playoffs each year. Typically playoffs only contain 8 teams but we felt as though that sample size would’ve been too small. Even with the additional teams, a single outlier from a team can drastically affect a region’s pattern.

Secondly, data completeness is also a limitation. The Kaggle data set we used explicitly notes that Chinese-hosted events lack API-accessible post-game statistics which potentially introduces bias. This limits our ability to confidently display China’s regional patterns. This affected our data because we had to remove 2 teams from China in order to fit our multilinear regression model due to their average team headshot rates being unavailable. After dropping these teams, there were only 9 America teams, 7 EMEA teams, 7 Pacific teams, and 6 China teams represented in the final regression model.

Third, our analysis is cross-sectional rather than longitudinal. This means that we only used data from a single point in time (VCT 2024) rather than comparing data over time. We chose to only capture the 2024 playoffs despite using a Kaggle set that contained all the VCT years due to the introduction of new agents and any game balance changes that would drastically change the pick rates of agents. By doing so, we cannot determine whether or not these regional differences persist across the entirety of VCT or the entire playerbase. The game constantly changes as new patches and agent updates roll out, making the predictors’ strength shift over time. 

Fourth, our research question only examines headshot percentage and agent pick rates when there are a plethora of other features that determine the outcome of a game. Some important factors that we left out were: economy management, clutch success rate, first-blood statistics, and team coordination (harder to measure). All of these factors are likely to influence the win rate of a game and may explain the variation in our data.
